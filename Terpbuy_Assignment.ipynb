# Write a query to show the quantity of items sold by each department. Sort the results by department name.


```python
df_quantity=pd.read_sql('Select department.department_name, order_line.quantity_sold\
                         from department, product Inner Join order_line on order_line.product_id= product.product_id \
                         group by department_name order by department_name asc;',conn);
df_quantity
```

# Using the query you wrote in Question 1, create a data visualization (e.g., a bar chart) showing all departments and the number of items each of them sold. Using a markdown cell, explain what you observe from the analysis.

```python
import matplotlib.pyplot as plt
```

```python
plt.bar("department_name", "quantity_sold", data = df_quantity, color = "blue")
plt.xlabel("department_name")
plt.xticks(rotation = 90)
plt.ylabel("Quantity")
plt.title("Quantity of items sold by each department")
plt.show()
```

By this analysis we can find out that all the departments in the Terpbuy have sold only one item from their department

# Write a query to show the number of orders placed in each year in which at least one order was placed. Hint: Search online to learn about MySQLâ€™s YEAR() function to query data.

```python
df_year=pd.read_sql('Select Year(orders.order_date), COUNT(order_line.quantity_sold)from orders, order_line \
                          GROUP BY year(orders.order_date) order by order_line.quantity_sold;',conn);
df_year
```
